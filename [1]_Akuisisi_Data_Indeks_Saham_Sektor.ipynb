```python
import os
import re
import pandas as pd
from glob import glob

# Path folder yang berisi file
folder_path = "/content/drive/MyDrive/Skripsi/Saham Sektor"

# Ambil semua file Excel dalam folder
file_list = sorted(glob(os.path.join(folder_path, "Ringkasan Indeks-*.xlsx")))

# Dictionary untuk menyimpan DataFrame setiap file berdasarkan tanggal
df_per_tanggal = {}

# Dictionary untuk menyimpan data sektor
sektor_dict = {
    'IDXENERGY': [], 'IDXBASIC': [], 'IDXINDUST': [], 'IDXNONCYC': [],
    'IDXCYCLIC': [], 'IDXHEALTH': [], 'IDXFINANCE': [], 'IDXPROPERT': [],
    'IDXTECHNO': [], 'IDXINFRA': [], 'IDXTRANS': []
}

# Loop melalui semua file Excel
for file in file_list:
    # Ekstrak tanggal dari nama file menggunakan regex
    match = re.search(r"(\d{8})", file)
    if not match:
        print(f"Skipping file (format tidak sesuai): {file}")
        continue

    tanggal_str = match.group(1)  # Contoh: "20210125"
    tanggal = pd.to_datetime(tanggal_str, format="%Y%m%d")  # Konversi ke datetime

    # Baca file Excel
    df = pd.read_excel(file)

    # Simpan DataFrame berdasarkan tanggalnya
    df_per_tanggal[tanggal_str] = df

    # Loop untuk memisahkan data berdasarkan sektor
    for sektor in sektor_dict.keys():
        sektor_data = df[df["Kode Indeks"] == sektor].copy()
        if not sektor_data.empty:
            sektor_data.insert(0, "Tanggal", tanggal)  # Tambahkan kolom tanggal
            sektor_dict[sektor].append(sektor_data)

# Gabungkan data per sektor menjadi satu DataFrame time series
for sektor in sektor_dict.keys():
    sektor_dict[sektor] = pd.concat(sektor_dict[sektor], ignore_index=True)

# Contoh akses DataFrame tertentu:
print("Data dari 25 Januari 2021:")
print(df_per_tanggal["20210125"].head())  # Menampilkan data dari file Ringkasan Indeks-20210125.xlsx

print("\nData sektor IDXENERGY (Time Series):")
print(sektor_dict["IDXENERGY"].head())
```

```python
import os
import hashlib
from glob import glob

# Path folder yang berisi file
folder_path = "/content/drive/MyDrive/Skripsi/Saham Sektor"

# Ambil semua file Excel dalam folder
file_list = sorted(glob(os.path.join(folder_path, "Ringkasan Indeks-*.xlsx")))

# Dictionary untuk menyimpan hash file dan daftar nama file yang sama
hash_dict = {}

# Fungsi untuk menghitung hash SHA-256 dari file
def hash_file(file_path):
    hasher = hashlib.sha256()
    with open(file_path, "rb") as f:
        while chunk := f.read(8192):
            hasher.update(chunk)
    return hasher.hexdigest()

# Loop melalui semua file untuk menghitung hash
for file in file_list:
    file_hash = hash_file(file)  # Hitung hash file

    # Simpan hash dalam dictionary
    if file_hash in hash_dict:
        hash_dict[file_hash].append(file)  # Tambahkan ke daftar duplikat
    else:
        hash_dict[file_hash] = [file]  # Simpan file pertama dengan hash ini

# Tampilkan file yang memiliki isi identik
duplikat_files = {h: files for h, files in hash_dict.items() if len(files) > 1}

if duplikat_files:
    print("🔍 File yang memiliki isi **identik** meski namanya berbeda:\n")
    for file_list in duplikat_files.values():
        print("\n".join(file_list))
        print("-" * 80)
else:
    print("✅ Tidak ada file duplikat berdasarkan isi.")

```

```python
import yfinance as yf
import pandas as pd

# Ambil data saham BBCA dari yfinance
bbca = yf.download("BBCA.JK", start="2021-01-25", end="2025-01-31")

# Reset index agar kolom 'Date' menjadi kolom biasa
bbca.reset_index(inplace=True)

# Pastikan 'Date' di BBCA dalam format datetime
bbca["Date"] = pd.to_datetime(bbca["Date"])

# Dictionary untuk menyimpan hasil tanggal yang hilang per sektor
hasil_tanggal_hilang = {}

# Loop melalui setiap sektor untuk mencocokkan tanggal
for sektor, df_sektor in sektor_dict.items():
    # Pastikan kolom 'Tanggal' di setiap sektor dalam format datetime
    df_sektor["Tanggal"] = pd.to_datetime(df_sektor["Tanggal"])

    # Ambil daftar tanggal unik dari sektor
    tanggal_sektor = set(df_sektor["Tanggal"])

    # Ambil daftar tanggal unik dari BBCA
    tanggal_bbca = set(bbca["Date"])

    # Cari tanggal yang ada di BBCA tetapi tidak ada di sektor
    tanggal_hilang = sorted(tanggal_bbca - tanggal_sektor)

    if tanggal_hilang:
        hasil_tanggal_hilang[sektor] = [tgl.strftime("%Y-%m-%d") for tgl in tanggal_hilang]

# Menampilkan hasil
if hasil_tanggal_hilang:
    print("🔍 **Tanggal yang ada di BBCA tetapi tidak ada di sektor:**\n")
    for sektor, tanggal in hasil_tanggal_hilang.items():
        print(f"📌 {sektor}: {', '.join(tanggal)}")
        print("-" * 50)
else:
    print("✅ Semua sektor memiliki tanggal yang sama dengan BBCA.")
```
