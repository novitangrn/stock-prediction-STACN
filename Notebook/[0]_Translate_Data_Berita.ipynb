```
from googletrans import Translator

data_indonesia = [
    "Awas Bubble! Saham Tesla Sudah Kemahalan",
    "Emas Boleh Cuan 25%, Tapi Ada yang Lebih 'Gila' Lho!",
    "Wall Street 2020: Dari Tekor Jadi Rekor!",
    "AS Depak 3 Emiten Telco China dari Wall Street, Ada Apa?",
    "Andai Tak Libur Tahun Baru, Rupiah Pasti Menguat!"
]

# Inisialisasi penerjemah
translator = Translator()

# Menerjemahkan setiap teks dalam daftar
data_english = [translator.translate(text, src='id', dest='en').text for text in data_indonesia]

# Menampilkan hasil terjemahan
for ind, eng in zip(data_indonesia, data_english):
    print(f"Original: {ind}")
    print(f"Translated: {eng}")
    print("-" * 50)
```

```
import pandas as pd

url = "https://raw.githubusercontent.com/novitangrn/dataset/refs/heads/main/Skripsi/Data%20Teks/news_kontan.csv"
df = pd.read_csv(url)
df['Tanggal'] = pd.to_datetime(df['Tanggal'], errors='coerce')
df
```

```
# Konversi kolom 'Tanggal' ke format datetime (tanpa jam)
df["Tanggal"] = pd.to_datetime(df["Tanggal"]).dt.date

# Grouping berdasarkan tanggal unik
df_grouped = df.groupby("Tanggal").size().reset_index(name="Jumlah_Record")

# Buat rentang tanggal yang lengkap
date_range = pd.date_range(start="2021-01-25", end="2025-01-31").date

# Cari tanggal yang tidak ada dalam df
missing_dates = set(date_range) - set(df_grouped["Tanggal"])

# Output hasil
print(f"Tanggal yang tidak ada dalam DataFrame: {sorted(missing_dates)}")
```
