# Classifying Kontan News

```python
from flashtext import KeywordProcessor
import pandas as pd
import string

# Fungsi Hapus Tanda Baca
def remove_punctuation(text):
    return text.translate(str.maketrans('', '', string.punctuation))

def fast_classify(df, column, dictionary, lowercase=False, case_sensitive=True):
    keyword_processor = KeywordProcessor()

    # Konfigurasi case sensitivity
    keyword_processor.case_sensitive = case_sensitive

    for sector, keywords in dictionary.items():
        keyword_processor.add_keywords_from_list(keywords)

    def get_sector_and_keywords(row):
        text = row[column]

        if not isinstance(text, str) or text.strip() == "":
            return "Unknown", []

        # Matching dengan atau tanpa lowercase sesuai kebutuhan
        content_to_match = text if case_sensitive else text.lower()
        matched_keywords = keyword_processor.extract_keywords(content_to_match)

        for keyword in matched_keywords:
            for sector, keywords in dictionary.items():
                if keyword in keywords:
                    return sector, matched_keywords

        return "Unknown", matched_keywords

    results = df.apply(get_sector_and_keywords, axis=1)
    matched_sectors = results.apply(lambda x: x[0])
    matched_keywords = results.apply(lambda x: x[1])

    return matched_sectors, matched_keywords


# --- Eksekusi Langkah-Langkah ---

# 1. Proses Judul (hapus tanda baca, case-sensitive, dict_saham)
all_data['Judul'] = all_data['Judul'].apply(remove_punctuation)
sectors, keywords = fast_classify(all_data, 'Judul', dict_saham, lowercase=False, case_sensitive=True)
all_data['Sektor'] = sectors
all_data['Matched Keywords'] = keywords

# 4. Proses Judul (case-sensitive, dict_all_sectors, dict_internasional, dict_obligasi untuk sektor 'Unknown')
unknown_data = all_data[all_data['Sektor'] == 'Unknown']
dict_combined = {**dict_all_sectors, **dict_internasional, **dict_obligasi}
sectors, keywords = fast_classify(unknown_data, 'Judul', dict_combined, lowercase=True, case_sensitive=False)

# Klasifikasi ke dalam 'All Sector', 'Internasional', atau 'Obligasi'
for idx, sector in zip(unknown_data.index, sectors):
    if sector in dict_all_sectors:
        all_data.at[idx, 'Sektor'] = 'All Sector'
    elif sector in dict_internasional:
        all_data.at[idx, 'Sektor'] = 'Internasional'
    elif sector in dict_obligasi:
        all_data.at[idx, 'Sektor'] = 'Obligasi'

all_data.loc[unknown_data.index, 'Matched Keywords'] = keywords

# 3. Proses Judul (lowercase, merged_dict untuk sektor 'Unknown')
unknown_data = all_data[all_data['Sektor'] == 'Unknown']
sectors, keywords = fast_classify(unknown_data, 'Judul', merged_dict, lowercase=True, case_sensitive=False)
all_data.loc[unknown_data.index, 'Sektor'] = sectors
all_data.loc[unknown_data.index, 'Matched Keywords'] = keywords
```
